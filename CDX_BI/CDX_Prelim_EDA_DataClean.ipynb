```python
#Create dict for surevys and doctors
#list of survey ID in billing table
surveys = billing['SurveyID'].unique()

#create dict for survey names and IDs from lusurvey
survey_dict = dict(zip(lusurvey.ID, lusurvey.Name))
```

```python
#create new df from billing
df = billing
```

```python
#add SurveyName column to df from survey_dict based on SurveyID
df['SurveyName'] = df['SurveyID'].map(survey_dict)
#add OfficeName column to df from office based on OfficeID
df['OfficeName'] = df['OfficeID'].map(office.set_index('ID')['LocalName'])
#add DoctorName column to df from doctor based on DoctorID
df['DoctorName'] = df['DoctorID'].map(doctor.set_index('ID')['LName'])
#add doctor address column to df from doctor based on DoctorID
df['DoctorAddress'] = df['DoctorID'].map(doctor.set_index('ID')['Address'])
```

```python
#new df from df with unique doctor IDs
df_docs = df.drop_duplicates(subset=['DoctorID'])
```

```python
from geopy.geocoders import Nominatim

# Function to geocode an address
def geocode_address(address):
    geolocator = Nominatim(user_agent="geocoder")
    location = geolocator.geocode(address)
    if location:
        return location.latitude, location.longitude
    else:
        return None, None

# Assuming you have a pandas DataFrame 'df' with an 'Address' column
# Create new 'Latitude' and 'Longitude' columns with geocoded coordinates
df_docs['Latitude'], df_docs['Longitude'] = zip(*df_docs['DoctorAddress'].apply(geocode_address))

# Print the DataFrame with geocoded coordinates
df_docs.head()
```

```python
#export df_docs to csv
df_docs.to_csv('recent_docs.csv')
```
